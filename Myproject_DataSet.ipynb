```python
##Get the year with the most spills
year_with_most_spills = None
max_spills = 0
for year in df1['Accident Year'].unique():
    spills_in_year = len(df1[df1['Accident Year'] == year])
    if spills_in_year > max_spills:
        max_spills = spills_in_year
        year_with_most_spills = year

# Display the year with the most spills
if year_with_most_spills is not None:
    print(f"The year with the most spills is {year_with_most_spills}.")
else:
    print("No spills found in the dataset.")
```
